```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2D

```python
x = np.array([3, 4])
```

```python
y = np.array([
    [1, 1],
    [0, 1]
])

shear = y @ x
```

```python
rotation_45 = np.pi / 4

y = np.array([
    [np.cos(rotation_45), -np.sin(rotation_45)],
    [np.sin(rotation_45), np.cos(rotation_45)]
])

rotation_45 = y @ x
```

```python
rotation_90 = np.pi / 2

y = np.array([
    [np.cos(rotation_90), -np.sin(rotation_90)],
    [np.sin(rotation_90), np.cos(rotation_90)]
])

rotation_90 = y @ x
```

```python
y = np.array([
    [2, 0],
    [0, 0.5]
])

scale = y @ x
```

```python
rotation_180 = np.pi

y = np.array([
    [np.cos(rotation_180), -np.sin(rotation_180)],
    [np.sin(rotation_180), np.cos(rotation_180)]
])

rotation_180 = y @ x
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

ax.quiver(0, 0, x[0], x[1], color='blue', angles='xy', scale_units='xy', scale=1, label='Vetor x (Original)', alpha=0.6)
ax.quiver(0, 0, shear[0], shear[1], color='red', angles='xy', scale_units='xy', scale=1, label='Shear', alpha=0.6)
ax.quiver(0, 0, rotation_45[0], rotation_45[1], color='green', angles='xy', scale_units='xy', scale=1, label='45 rotation', alpha=0.6)
ax.quiver(0, 0, rotation_90[0], rotation_90[1], color='orange', angles='xy', scale_units='xy', scale=1, label='90 rotation', alpha=0.6)
ax.quiver(0, 0, scale[0], scale[1], color='purple', angles='xy', scale_units='xy', scale=1, label='Scale', alpha=0.6)
ax.quiver(0, 0, rotation_180[0], rotation_180[1], color='black', angles='xy', scale_units='xy', scale=1, label='180 rotation', alpha=0.6)

limite = max(np.max(np.abs(x)), np.max(np.abs(shear))) + 1
ax.set_xlim([-limite, limite])
ax.set_ylim([-limite, limite])

ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Transformao Linear no Espao 2D')
ax.grid(True)
ax.legend()
plt.show()
```

## 3D

```python
x_3d = np.array([3, 4, 1])
```

```python
y = np.array([
    [1, 1, 0],
    [0, 1, 0],
    [1, 1, 1]
])

transformation = y @ x_3d
```

```python
rotation_45 = np.pi / 4

y = np.array([
    [np.cos(rotation_45), -np.sin(rotation_45), 0],
    [np.sin(rotation_45), np.cos(rotation_45), 0],
    [0, 0, 1]
])

rotated_45 = y @ x_3d
```

```python
rotation_90 = np.pi / 2

y = np.array([
    [np.cos(rotation_90), -np.sin(rotation_90), 0],
    [np.sin(rotation_90), np.cos(rotation_90), 0],
    [0, 0, 1]
])

rotated_90 = y @ x_3d
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.quiver(0, 0, 0, x_3d[0], x_3d[1], x_3d[2], color="red", label='Original')
ax.quiver(0, 0, 0, transformation[0], transformation[1], transformation[2], color="blue", label='Transformation')
ax.quiver(0, 0, 0, rotated_45[0], rotated_45[1], rotated_45[2], color="green", label='45 rotation')
ax.quiver(0, 0, 0, rotated_90[0], rotated_90[1], rotated_90[2], color="orange", label='90 rotation')


limite = max(np.max(np.abs(x)), np.max(np.abs(transformation))) + 1
ax.set_xlim([-limite, limite])
ax.set_ylim([-limite, limite])
ax.set_zlim([-limite, limite])

ax.legend()
plt.show()
```
