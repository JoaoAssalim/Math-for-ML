```python
import numpy as np
```

```python
X = np.random.rand(5, 5)
X
```

```python
X_mean = X.mean(axis=0)
X_mean
```

```python
X_minus_mean = X - X_mean
X_minus_mean
```

```python
C = (X_minus_mean.T @ X_minus_mean) / (X.shape[0] -1)
C
```

```python
eigenval, eigenvec = np.linalg.eig(C)

idx = np.argsort(eigenval)[::-1]

eigenval_sorted = eigenval[idx]
eigenvec_sorted = eigenvec[:, idx]


print(f"EigenValues: {eigenval_sorted}\n\nEigenVectors: {eigenvec_sorted}")
```

```python
top_eigenvals = eigenval_sorted[:2]
top_eigenvecs = eigenvec_sorted[:, :2]
```

```python
top_eigenvecs
```

```python
Xpca = X_minus_mean @ top_eigenvecs
```
