```python
import numpy as np
```

```python
def dot_product(u, v):
    # sum of the products of their corresponding components
    return u @ v
```

```python
def norm(v):
    # magnitude of a vector
    return sum([component ** 2 for component in v]) ** 0.5
```

```python
def similarity_score(u, v):
    dot_prod = dot_product(u, v)

    norm_u = norm(u)
    norm_v = norm(v)

    return dot_prod / (norm_u * norm_v)
```

```python
# [Action, Romance, Drama]
filme_A = np.array([5, 0, 1])
filme_B = np.array([4, 1, 1])
filme_C = np.array([0, 5, 1])

sim_AB = similarity_score(filme_A, filme_B)
sim_AC = similarity_score(filme_A, filme_C)

# Close to 1 = Very Similar
# Close to 0 = Very Different
print(f"Similarity A and B: {sim_AB:.4f}") # Very similar
print(f"Similarity A and C: {sim_AC:.4f}")
```
